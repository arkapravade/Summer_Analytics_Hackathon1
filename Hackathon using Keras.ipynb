```python
bottom98 = df.sort_values('price1',ascending=False).iloc[350:]
bottom98.head()
```

```python
#X = df[['Rating','actprice1']].values
#y = df['price1'].values

X = bottom98[['Rating','actprice1']].values
y = bottom98['price1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout  #randomly turns off layers
from tensorflow.keras.callbacks import EarlyStopping
#from keras.losses import mean_squared_error
```

```python
early_stop = EarlyStopping(monitor='val_loss', mode='min',verbose=1, patience=20)
```

```python
model = Sequential()

model.add(Dense(1500,activation='relu'))
#model.add(Dropout(0.5))

#model.add(Dense(700,activation='sigmoid'))

model.add(Dense(1500,activation='relu'))

#model.add(Dense(700,activation='sigmoid'))

#model.add(Dense(350,activation='relu'))
#model.add(Dropout(0.4))

#model.add(Dense(350,activation='relu'))

model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,epochs=2000,batch_size=128,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
dfloss = pd.DataFrame(model.history.history)
```

```python
dfloss.plot()
```

```python
predictions = model.predict(X_test)
#predictions
plt.figure(figsize=(12,8))
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,color='r')
```

```python
np.sqrt(model.evaluate(X_train,y_train))
```

```python
np.sqrt(model.evaluate(X_test,y_test))
```
